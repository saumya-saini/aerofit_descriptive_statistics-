2 Way - Contingency Table, Marginal Probability and Conditional Probability

Contingency Table 1

```python
# Product vs Gender (Contingency Table)
pd.crosstab(df['Product'], df['Gender'])
```

```python
# Marginal Probability: Product
df['Product'].value_counts(normalize=True)
```

KP281 has the highest purchase probability, indicating it is the most popular treadmill among customers.

```python
# Marginal Probability: Gender
df['Gender'].value_counts(normalize=True)
```

A higher proportion of treadmill buyers are male compared to female customers.

```python
# Conditional Probability: P(G|P) (Probability of Gender given Product)
pd.crosstab(df['Product'], df['Gender'], normalize = 'index')
```

The premium treadmill KP781 is predominantly purchased by male customers (82.5%), while KP281 shows an equal gender distribution. This suggests that high-end treadmill models may appeal more to male customers.

```python
# Conditional Probability: P(Product|Gender) (probability of product given gender)
pd.crosstab(df['Product'], df['Gender'], normalize = 'columns')
```

Females mostly buy entry-level treadmills.
Males buy premium treadmills much more often.

Business Insights:

Insight 1

KP281 is the most purchased treadmill, indicating it appeals to a broader customer segment.

Insight 2

Male customers dominate treadmill purchases, suggesting Aerofit products may currently attract more male buyers.

Insight 3

A higher proportion of KP781 buyers are male, indicating the premium treadmill is more popular among male customers.

Insight 4

Aerofit could design targeted marketing strategies to increase female participation in premium treadmill purchases.

Contingency Table 2

```python
# Product vs Marital Status
pd.crosstab(df['Product'], df['MaritalStatus'])
```

```python
# Marginal Probability: Product
df['Product'].value_counts(normalize=True)
```

KP281 has the highest purchase probability, indicating it is the most popular treadmill among customers.

```python
# Marginal Probability: Marital Status
df['MaritalStatus'].value_counts(normalize=True)
```

A higher proportion of treadmill buyers are partnered customers (59.4%) compared to single customers (40.5%)

```python
# Conditional Probabilities: P(Marital Status | Product) (probability of marital status given product)
pd.crosstab(df['Product'], df['MaritalStatus'], normalize='index')
```

A larger proportion of KP281 and KP481 buyers are partnered customers, suggesting these treadmills may appeal more to households.

```python
# Conditional Proabability: P(Product | MaritalStatus) (probability of product given marital status)
pd.crosstab(df['Product'], df['MaritalStatus'], normalize='columns')
```

Both partnered and single customers show a higher likelihood of purchasing KP281 compared to other treadmill models.

Business Insights:

Insight 1

KP281 remains the most popular treadmill across both partnered and single customers.

Insight 2

Partnered customers constitute a larger share of treadmill buyers, indicating that treadmills may often be purchased for household use.

Insight 3

The premium treadmill KP781 has a relatively smaller share of partnered buyers, suggesting it may appeal to a more niche customer segment.

Contingency Table 3

```python
# Product vs Fitness
pd.crosstab(df['Product'], df['Fitness'])
```

```python
# Marginal Probability: Product
df['Product'].value_counts(normalize=True)
```

KP281 has the highest purchase probability, indicating it is the most popular treadmill among customers.

```python
# Marginal Probability: Fitness
df['Fitness'].value_counts(normalize=True)
```

```python
# Conditional Probability: P(Fitness|Product) (probability of fitness given product)
pd.crosstab(df['Product'], df['Fitness'], normalize='index')
```

Customers purchasing KP781 tend to have higher fitness levels (4–5), while KP281 buyers mostly fall in moderate fitness levels (3).

```python
# Conditional Probability: P(Product|Fitness) (probability of product given fitness)
pd.crosstab(df['Product'], df['Fitness'], normalize='columns')
```

Business Insights:

Insight 1

Customers with higher fitness levels are more likely to purchase the premium treadmill KP781.

Insight 2

KP281 is primarily purchased by customers with moderate fitness levels, suggesting it is suitable for beginners or casual users.

Insight 3

The KP481 treadmill attracts customers with mid-range fitness levels, positioning it as a balanced option between entry-level and premium models.
